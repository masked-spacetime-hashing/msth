```python
import matplotlib.pyplot as plt
import numpy as np

sparse_funcs = {
        "logoneplus": lambda x: np.log(2 * (x**2) + 1.0),
        "l1": lambda x: np.abs(x),
}

x = np.linspace(0, 100, 1000)

l1_y = sparse_funcs["l1"](x)
logoneplus_y = sparse_funcs["logoneplus"](x)
plt.plot(x, l1_y)
plt.plot(x, logoneplus_y)
plt.show()
```
